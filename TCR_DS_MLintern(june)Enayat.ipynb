#Clean the Dataset

There is no missing value in this dataset. Nevertheless, there are values like “unknown”, “others”, which are helpless just like missing values. Thus, these ambiguous values are removed from the dataset.

```python
#Deleting the rows of column 'poutcome' if it contains 'other'
condition = dset.poutcome == 'other'
dset = dset.drop(dset[condition].index, axis = 0, inplace = False)
```

```python
#Replacing 'unknown' in job and education with 'other'
dset[['job','education']] = dset[['job','education']].replace(['unknown'],'other')
```

##Outliers Check
> In order to capture the general trend in the dataset, outliers in the column “balance” are dropped. Outliers are defined as the values which are more than three standard deviations away from the mean. In sum, 2556 rows of data were removed.

Drop outliers in the column 'balance'

```python
from scipy.stats import zscore

dset[['balance']].mean()
dset[['balance']].mean()

dset['balance_outliers'] = dset['balance']
dset['balance_outliers']= zscore(dset['balance_outliers'])

balance_outliers_condition = (dset['balance_outliers']>3) | (dset['balance_outliers']<-3 )
dset = dset.drop(dset[balance_outliers_condition].index, axis = 0, inplace = False)
```

```python
dset = dset.drop('balance_outliers', axis=1)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import f1_score, precision_score, recall_score, roc_auc_score
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(random_state=0)
logreg.fit(X_train, Y_train)
y_pred_lr = logreg.predict(X_test)
roc=roc_auc_score(Y_test, y_pred_lr)
acc = accuracy_score(Y_test, y_pred_lr)
prec = precision_score(Y_test, y_pred_lr)
rec = recall_score(Y_test, y_pred_lr)
f1 = f1_score(Y_test, y_pred_lr)


acc_logreg = round(accuracy_score(y_pred_lr, Y_test) * 100, 2)
results1 = pd.DataFrame([['Logistic Regression', acc, acc_logreg, prec,rec, f1,roc]],
               columns = ['Model', 'Accuracy', 'Accuracy %', 'Precision', 'Recall', 'F1 Score','ROC'])
results1
```

```python
# Logistic Regression
cm = confusion_matrix(Y_test, y_pred_lr)
print(cm)
sns.heatmap(cm,square=True, annot=True,linewidths=1,cmap="RdPu")
```

```python
# Gaussian Naive Bayes
from sklearn.naive_bayes import GaussianNB


gaussian = GaussianNB()
gaussian.fit(X_train, Y_train)
y_pred_gnb = gaussian.predict(X_test)
roc=roc_auc_score(Y_test, y_pred_gnb)
acc = accuracy_score(Y_test, y_pred_gnb)
prec = precision_score(Y_test, y_pred_gnb)
rec = recall_score(Y_test, y_pred_gnb)
f1 = f1_score(Y_test, y_pred_gnb)

acc_gaussian = round(accuracy_score(y_pred_gnb, Y_test) * 100, 2)
results2 = pd.DataFrame([['Gaussian Naive Bayes', acc, acc_gaussian, prec,rec, f1,roc]],
               columns = ['Model', 'Accuracy', 'Accuracy %', 'Precision', 'Recall', 'F1 Score','ROC'])
results2
```

```python
# Support Vector Machines
from sklearn.svm import SVC

svc = SVC(probability=True)
svc.fit(X_train, Y_train)
y_pred_svm = svc.predict(X_test)
roc=roc_auc_score(Y_test, y_pred_svm)
acc = accuracy_score(Y_test, y_pred_svm)
prec = precision_score(Y_test, y_pred_svm)
rec = recall_score(Y_test, y_pred_svm)
#f1 = f1_score(Y_test, y_pred_svm)
f1 = f1_score(Y_test, y_pred_svm, average='weighted', labels=np.unique(y_pred_svm)) 

acc_svm = round(accuracy_score(y_pred_svm, Y_test) * 100, 2)
results3 = pd.DataFrame([['Support Vector Machines', acc, acc_svm, prec,rec, f1,roc]],
               columns = ['Model', 'Accuracy', 'Accuracy %', 'Precision', 'Recall', 'F1 Score','ROC'])

results3
```

there is no F-score to calculate for this label, and thus the F-score for this case is considered to be 0.0. Since you requested an average of the score, you must take into account that a score of 0 was included in the calculation, and this is why scikit-learn is showing you that warning.
What you can do, is decide that you are not interested in the scores of labels that were not predicted, and then explicitly specify the labels you are interested in (which are labels that were predicted at least once):
